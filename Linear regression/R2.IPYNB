```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt 
```

```python
df = pd.DataFrame({'distance':[1,2,3,4,5],'price':[10,20,30,40,50]})
```

```python
X = df[['distance']]
y = df[['price']]
```

```python
model = LinearRegression()
model.fit(X,y)
```

```python
y_predict =  model.predict(X)
```

```python
plt.scatter('distance','price',data = df,label= 'Actual')
plt.plot(df['distance'],y_predict,color = 'red',label = 'Predicted')
plt.legend()
plt.plot()

```

```python
df = pd.DataFrame({'distance':[1,2,3,4,5],'price':[13,22,38,47,59]})
```

```python
X = df[['distance']]
y = df[['price']]
model = LinearRegression()
model.fit(X,y)
y_predict =  model.predict(X)
```

```python
plt.scatter('distance','price',data = df,label= 'Actual')
plt.scatter(df['distance'],y_predict,color = 'red',label = 'Predicted')
plt.legend()
plt.plot()

```

```python
model.score(X,y)
```

```python
y_actual = y
```

```python
sse = np.sum((y - y_predict)**2)
print(sse)
```

```python
sse_mean = np.sum((y- np.mean(y_actual))**2)
sse_mean
```
